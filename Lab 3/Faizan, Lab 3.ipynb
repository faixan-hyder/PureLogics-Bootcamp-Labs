```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('hourly_weather_10_days.csv')

# Step 1: Display first few rows to verify the load
print("--- First 5 rows of the dataset ---")
print(df.head())

# Step 1: Check the number of rows and columns
print(f"\nDataset shape (rows, columns): {df.shape}")
```

```python
# Step 2: Data ka health check-up
print("--- DataFrame Info ---")
print(df.info())

print("\n--- Descriptive Statistics ---")
print(df.describe())

print("\n--- Missing Value Count ---")
print(df.isna().sum())
```

```python
print(df.isnull().values.any())
```

```python
# Har column ka average nikal kar khali jagah bhar do
df_filled = df.fillna(df.mean(numeric_only=True))
print("Missing values ab fill ho gayi hain!")
```

```python
# Jin rows mein data nahi hai, unhein hata do
df_cleaned = df.dropna()
print("Rows jahan data nahi tha, unhein hata diya gaya hai.")
```

```python
# Temperature column ka mean (average) nikalna
temp_mean = df['temperature_C'].mean()

# Missing values ko mean se fill karna
df['temperature_C'] = df['temperature_C'].fillna(temp_mean)

# Verify karna ki kya ab koi missing value bachi hai
print(f"Ab 'temperature_C' mein missing values ki tadaad: {df['temperature_C'].isna().sum()}")
```

```python
# 'timestamp' column ko datetime format mein convert karna
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Verify karne ke liye info check karein
print(df.info())
```

```python
# Un columns ki list jinme missing values hain
cols_to_fill = ['humidity_%', 'wind_speed_kmph', 'pressure_hPa', 'visibility_km']

# Loop chala kar sabka mean nikal kar fill kar do
for col in cols_to_fill:
    col_mean = df[col].mean()
    df[col] = df[col].fillna(col_mean)

# Final check
print(df.info())
```

```python
import numpy as np

# Humne 'temperature_C' aur 'humidity_%' ko numpy arrays mein liya
temp_array = df['temperature_C'].to_numpy()
humidity_array = df['humidity_%'].to_numpy()

print("Temperature Array ka shape:", temp_array.shape)
print("Pehle 5 values:", temp_array[:5])
```

```python
# Step 5 ka pehla part: DataFrame se NumPy arrays mein convert karna
temp = df['temperature_C'].to_numpy()
wind = df['wind_speed_kmph'].to_numpy()

# Verify karne ke liye
print("Temp array shape:", temp.shape)
```

```python
# Matrix mein reshape karna
temp_matrix = temp.reshape((10, 24))

# Ab aap is matrix par min, max, aur mean nikal sakte hain
# axis=1 ka matlab hai ki hum har din (row) ka calculation kar rahe hain
daily_min = np.min(temp_matrix, axis=1)
daily_max = np.max(temp_matrix, axis=1)
daily_mean = np.mean(temp_matrix, axis=1)

print("Har din ka average temperature:", daily_mean)
```

```python
# Normalization: (Value - Mean) / Standard Deviation
# Numpy ka use karte huye
temp_mean = np.mean(temp_matrix)
temp_std = np.std(temp_matrix)

# Yeh formula poori matrix par ek saath apply hoga (Vectorization)
normalized_temp = (temp_matrix - temp_mean) / temp_std

print("Normalized Matrix ki pehli 2 rows:")
print(normalized_temp[:2])
```

```python
# 1. Mask banana (shart: wind speed > 15)
mask = wind > 15

# 2. Is mask ko original wind array par apply karna
high_wind = wind[mask]

# 3. Check karna ki kitne readings match hui
print(f"Total readings jahan wind speed 15 se zyada thi: {len(high_wind)}")
print("Pehle 5 high-wind values:", high_wind[:5])
```

```python
def daily_summary(matrix):
    # 'summaries' list mein har din ka dict store karenge
    summaries = []
    for i in range(matrix.shape[0]):
        day_data = matrix[i]
        day_stats = {
            'day': i + 1,
            'min': np.min(day_data),
            'max': np.max(day_data),
            'mean': np.mean(day_data)
        }
        summaries.append(day_stats)
    return summaries

# Function call
results = daily_summary(temp_matrix)

# Print results
for res in results:
    print(res)
```

```python
import pandas as pd
summary_df = pd.DataFrame(daily_summary(temp_matrix))
print(summary_df)
```
