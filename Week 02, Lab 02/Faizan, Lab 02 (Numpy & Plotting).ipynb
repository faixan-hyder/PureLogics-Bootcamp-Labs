```python
U, s, V = np.linalg.svd(A)
print(U.shape, V.shape, s.shape)
```

```python
# Task 6

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

data = np.random.rand(50)
print("Data array (50 random numbers):")
print(data)
print(f"\nShape: {data.shape}")
print(f"Data type: {data.dtype}")

mean_val = np.mean(data)
median_val = np.median(data)
std_val = np.std(data)

print("\nStatistics:")
print("-" * 30)
print(f"Mean:     {mean_val:.4f}")
print(f"Median:   {median_val:.4f}")
print(f"Standard Deviation: {std_val:.4f}")

min_val = np.min(data)
max_val = np.max(data)
range_val = max_val - min_val
sum_val = np.sum(data)
var_val = np.var(data)

print("\nAdditional Statistics:")
print("-" * 30)
print(f"Minimum:  {min_val:.4f}")
print(f"Maximum:  {max_val:.4f}")
print(f"Range:    {range_val:.4f}")
print(f"Sum:      {sum_val:.4f}")
print(f"Variance: {var_val:.4f}")
print(f"Count:    {len(data)}")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=10, color='blue', edgecolor='black', alpha=0.7)
plt.axvline(mean_val, color='red', linewidth=2, label=f'Mean: {mean_val:.3f}')
plt.axvline(median_val, color='green', linewidth=2, label=f'Median: {median_val:.3f}')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram with Mean & Median')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)

plt.subplot(1, 2, 2)
plt.boxplot(data, vert=True)
plt.title('Box Plot')
plt.ylabel('Values')
plt.grid(True, linestyle='--', alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Task 7

import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

categories = ['Math', 'Science', 'History', 'English']
values = np.random.randint(10, 50, 4)

print("Categories:", categories)
print("Values:", values)
print("Total:", sum(values))
print()

plt.figure(figsize=(10, 8))

colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFA07A']

explode = (0.05, 0.05, 0.05, 0.05)

plt.pie(values, 
        labels=categories,
        autopct='%1.1f%%',
        startangle=90,
        colors=colors,
        explode=explode,
        shadow=True,
        textprops={'fontsize': 12},
        wedgeprops={'edgecolor': 'black', 'linewidth': 1})

plt.title('Subject Distribution', fontsize=16, fontweight='bold')
plt.axis('equal')

plt.text(0, -1.2, f'Total Students: {sum(values)}', 
         fontsize=12, ha='center', va='center')

plt.show()

print("\nSummary:")
print("-" * 30)
for category, value in zip(categories, values):
    percentage = (value / sum(values)) * 100
    print(f"{category:10s}: {value:3d} students ({percentage:5.1f}%)")
```

```python
# Task 8

import numpy as np

matrix = np.random.randint(1, 11, size=(3, 3))
print("Original Matrix (3×3):")
print(matrix)
print(f"\nShape: {matrix.shape}")

transposed = matrix.T
print("\nTransposed Matrix (3×3):")
print(transposed)
print(f"Shape: {transposed.shape}")

col_sums = np.sum(transposed, axis=0)
print("\nSum of each column (of transposed):")
print(col_sums)

print("\n--- Verification ---")
for col in range(3):
    column_values = transposed[:, col]
    col_sum = np.sum(column_values)
    print(f"Column {col}: {column_values} → Sum = {col_sum}")

row_sums = np.sum(transposed, axis=1)
print(f"\nSum of each row (of transposed): {row_sums}")

original_row_sums = np.sum(matrix, axis=1)
print(f"\nRow sums of original: {original_row_sums}")
print("Equal to column sums of transposed?", np.array_equal(col_sums, original_row_sums))
```

```python
# Task 9

import numpy as np

def square_array(arr):
    """
    Squares every element in the array.
    
    Parameters:
    arr: numpy array
    
    Returns:
    numpy array with all elements squared
    """
    return arr ** 2

print("--- Testing square_array function ---\n")

arr1 = np.array([1, 2, 3, 4, 5])
print("1D Array:")
print(f"Original: {arr1}")
print(f"Squared: {square_array(arr1)}\n")

arr2 = np.array([[1, 2, 3],
                 [4, 5, 6]])
print("2D Array:")
print(f"Original:\n{arr2}")
print(f"Squared:\n{square_array(arr2)}\n")

arr3 = np.array([-3, -2, -1, 0, 1, 2, 3])
print("Negative Numbers:")
print(f"Original: {arr3}")
print(f"Squared: {square_array(arr3)}\n")

arr4 = np.array([1.5, 2.5, 3.5, 4.5])
print("Float Numbers:")
print(f"Original: {arr4}")
print(f"Squared: {square_array(arr4)}\n")

original = np.array([1, 2, 3])
squared = square_array(original)
print("Function doesn't modify original:")
print(f"Original: {original}")
print(f"Squared: {squared}")
print(f"Original still: {original}")  # Unchanged
```

```python
# Task 10

import numpy as np

array = np.random.rand(5, 5)
print("Original array:")
print(array)
print(f"\nShape: {array.shape}")

array[0, 0] = np.nan
array[1, 2] = np.nan
array[2, 4] = np.nan
array[3, 1] = np.nan
array[4, 3] = np.nan

print("\nArray with NaN values:")
print(array)
print(f"Number of NaN values: {np.isnan(array).sum()}")

col_means = np.nanmean(array, axis=0)
print("\nColumn means (ignoring NaN):")
print(col_means)

array_filled = array.copy()
for col in range(array_filled.shape[1]):
    col_mean = col_means[col]
    array_filled[:, col] = np.where(np.isnan(array_filled[:, col]), 
                                   col_mean, 
                                   array_filled[:, col])

print("\nArray after filling NaN:")
print(array_filled)
print(f"NaN remaining: {np.isnan(array_filled).sum()}")

# Step 5: Verification
print("\n--- Verification ---")
for col in range(array_filled.shape[1]):
    column = array_filled[:, col]
    print(f"Column {col}: Mean = {np.mean(column):.4f}, No NaN? {np.all(~np.isnan(column))}")
```
