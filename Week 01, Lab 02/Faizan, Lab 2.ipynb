```python
faizan_text = "Faizan is a coder. Faizan is learning Python. Faizan is dedicated."
words = faizan_text.replace('.', '').split()
word_freq = {}
for word in words:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1
print(f"Word frequency for {word_freq}")

count_faizan = words.count('Faizan')
print(f"Faizan sentence mein {count_faizan} baar aaya hai.")

faizan_only = []
for word in words:
    if word == "Faizan":
        faizan_only.append(word)
print(faizan_only)
```
