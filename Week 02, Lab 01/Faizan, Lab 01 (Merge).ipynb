```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["Faisalabad","Lahore","Islamabad"],
    "temperature": [10,12,14],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["Faisalabad","Lahore","Islamabad"],
    "humidity": [65,68,75],
})
df2
```

```python
df3 = pd.merge(df1, df2, on="city")
df3
```

```python
df1 = pd.DataFrame({
    "city": ["Multan","Peshawar","Gujranwala", "Sheikhupura"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["Multan","Peshawar","Jaranwala"],
    "humidity": [65,68,71],
})
df2
```

```python
df3=pd.merge(df1,df2,on="city",how="inner")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="left")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="right")
df3
```

```python
df3=pd.merge(df1,df2,on="city",how="outer",indicator=True)
df3
```

```python
df1 = pd.DataFrame({
    "city": ["FSD","LHR","KHR","ISB"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["FSD","LHR","RWL"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second'))
df3
```

```python
df1 = pd.DataFrame({
    "city": ["FSD","LHR","ISB"],
    "temperature": [21,14,35],
})
df1.set_index('city',inplace=True)
df1
```

```python
df2 = pd.DataFrame({
    "city": ["LHR","FSD","ISB"],
    "humidity": [65,68,75],
})
df2.set_index('city',inplace=True)
df2
```

```python
df1.join(df2,lsuffix='_l', rsuffix='_r')
```
