```python
# ELBOW METHOD - SIMPLE SOLUTION
# Find Optimal Number of Clusters

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import pandas as pd
import warnings
import os

# Fix warnings
os.environ["OMP_NUM_THREADS"] = "1"
warnings.filterwarnings('ignore')

iris = load_iris()
df = pd.DataFrame(iris.data)

print("Data loaded!")
print(f"Shape: {df.shape}")
print(f"Features: {iris.feature_names}\n")

# CALCULATE SSE FOR DIFFERENT K VALUES
sse = []  # Sum of Squared Errors

# Try K from 1 to 10
for k in range(1, 11):
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    km.fit(df)
    sse.append(km.inertia_)  # inertia_ = SSE
    print(f"K = {k}: SSE = {km.inertia_:.2f}")

# PLOT ELBOW CURVE
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o', linestyle='-', color='blue', linewidth=2, markersize=8)
plt.scatter(range(1, 11), sse, color='red', s=50)

# Add labels
plt.xlabel('Number of Clusters (K)', fontsize=12)
plt.ylabel('SSE (Sum of Squared Errors)', fontsize=12)
plt.title('Elbow Method - Optimal K', fontsize=14, fontweight='bold')

# Add grid
plt.grid(True, alpha=0.3)

# Mark elbow point
plt.axvline(x=3, color='red', linestyle='--', alpha=0.7, label='Elbow Point (K=3)')
plt.legend()

plt.show()

# FIND OPTIMAL K
# Elbow point = where SSE decrease slows down
# For Iris, K=3 is optimal (matches 3 species)

optimal_k = 3
print(f"\nOptimal K = {optimal_k} (Elbow Point)")

# ---- APPLY K-MEANS WITH OPTIMAL K ----
print("APPLYING K-MEANS WITH OPTIMAL K = 3")

# Create model with K=3
model = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
y_predicted = model.fit_predict(df)

# Add predictions to dataframe
df['Cluster'] = y_predicted

# Count points in each cluster
print(f"\nCluster Distribution:")
for i in range(optimal_k):
    count = sum(y_predicted == i)
    print(f"   Cluster {i}: {count} points")

# VISUALIZE CLUSTERS 
plt.figure(figsize=(12, 5))

# Plot 1: Scatter plot with clusters
plt.subplot(1, 2, 1)
colors = ['red', 'green', 'blue']
markers = ['.', '+', 'x']

for i in range(optimal_k):
    cluster_data = df[df['Cluster'] == i]
    plt.scatter(cluster_data[2], cluster_data[3], 
               color=colors[i], marker=markers[i], 
               label=f'Cluster {i}', s=30)

plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title(f'K-Means Clustering (K={optimal_k})')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 2: Elbow curve with optimal K marked
plt.subplot(1, 2, 2)
plt.plot(range(1, 11), sse, marker='o', color='blue', linewidth=2)
plt.scatter(range(1, 11), sse, color='red', s=50)
plt.axvline(x=optimal_k, color='red', linestyle='--', linewidth=2, label=f'Optimal K={optimal_k}')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('SSE')
plt.title('Elbow Curve with Optimal K')
plt.grid(True, alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()

# FINAL SUMMARY 
print("FINAL SUMMARY")
print(f"Total samples: {len(df)}")
print(f"Optimal clusters: {optimal_k}")
print(f"Features used: {len(df.columns)-1}")
print(f"\nK-Means with K={optimal_k} successfully applied!")

# Compare with actual labels
actual_labels = iris.target
print(f"\nActual Iris species: 3 (Setosa, Versicolor, Virginica)")
print(f"K-Means clusters: {optimal_k}")
print(f"\nMatch! K=3 matches the 3 species!")
```
