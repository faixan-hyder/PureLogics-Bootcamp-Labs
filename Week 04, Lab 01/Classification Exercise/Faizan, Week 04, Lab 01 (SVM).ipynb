```python
# SVM - TEST SIZE VARIATION EXPERIMENT
# See how test size affects accuracy

from sklearn.datasets import load_digits
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import pandas as pd
import matplotlib.pyplot as plt

# LOAD DATA 
digits = load_digits()
X = digits.data
y = digits.target

print("SVM TEST SIZE EXPERIMENT")
print(f"Total samples: {len(X)}")
print(f"Features: {X.shape[1]}\n")

# TRY DIFFERENT TEST SIZES 
test_sizes = [0.10, 0.20, 0.30, 0.40, 0.50]
results = []

for test_size in test_sizes:
    print(f"\nTest Size = {test_size*100:.0f}%")
    
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=test_size, random_state=42
    )
    
    print(f"   Train: {len(X_train)} samples")
    print(f"   Test:  {len(X_test)} samples")
    
    # Train SVM
    model = SVC()
    model.fit(X_train, y_train)
    
    # Calculate accuracy
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)
    
    print(f"   Train Accuracy: {train_score:.4f}")
    print(f"   Test Accuracy:  {test_score:.4f}")
    print(f"   Difference:     {train_score - test_score:.4f}")
    
    # Store results
    results.append({
        'test_size': test_size,
        'train_samples': len(X_train),
        'test_samples': len(X_test),
        'train_acc': train_score,
        'test_acc': test_score,
        'diff': train_score - test_score
    })

# COMPARE RESULTS
print("COMPARISON TABLE")

for r in results:
    print(f"Test {r['test_size']*100:.0f}%: Train={r['train_acc']:.4f}, Test={r['test_acc']:.4f}, Diff={r['diff']:.4f}")

# FIND BEST
best = max(results, key=lambda x: x['test_acc'])
print(f"\nBest Test Size: {best['test_size']*100:.0f}% with accuracy {best['test_acc']:.4f}")

# PLOT RESULTS
plt.figure(figsize=(12, 5))

# Plot 1: Accuracy vs Test Size
plt.subplot(1, 2, 1)
test_sizes_plot = [r['test_size']*100 for r in results]
train_acc = [r['train_acc'] for r in results]
test_acc = [r['test_acc'] for r in results]

plt.plot(test_sizes_plot, train_acc, 'bo-', label='Train Accuracy', linewidth=2, markersize=8)
plt.plot(test_sizes_plot, test_acc, 'ro-', label='Test Accuracy', linewidth=2, markersize=8)
plt.xlabel('Test Size (%)')
plt.ylabel('Accuracy')
plt.title('Accuracy vs Test Size')
plt.legend()
plt.grid(True, alpha=0.3)

# Plot 2: Samples distribution
plt.subplot(1, 2, 2)
train_samples = [r['train_samples'] for r in results]
test_samples = [r['test_samples'] for r in results]
x_pos = range(len(test_sizes))

plt.bar(x_pos, train_samples, label='Train', color='blue', alpha=0.7)
plt.bar(x_pos, test_samples, label='Test', color='red', alpha=0.7, bottom=train_samples)
plt.xlabel('Test Size (%)')
plt.ylabel('Number of Samples')
plt.title('Train/Test Split Distribution')
plt.xticks(x_pos, [f'{s*100:.0f}%' for s in test_sizes])
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\nExperiment complete!")
```
