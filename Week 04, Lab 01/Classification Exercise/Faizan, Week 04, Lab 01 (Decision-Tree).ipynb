```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.model_selection import train_test_split
import math
```

```python
df = pd.read_csv("Copy of titanic.csv")
df
```

```python
# Drop only completely useless columns
df = df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], axis='columns')

print("Columns dropped!")
print(f"Remaining columns: {list(df.columns)}")
df.head()
```

```python
# Age - fill with mean
mean_age = math.floor(df.Age.mean())
df['Age'] = df.Age.fillna(mean_age)
print(f"Age filled with: {mean_age}")

# Embarked - fill with most common value ('S')
df['Embarked'] = df.Embarked.fillna('S')
print("Embarked filled with: S (most common)")

# Check missing values
print("\nMissing values check:")
print(df.isnull().sum())
```

```python
# Encode Sex
LE_sex = LabelEncoder()
df['Sex'] = LE_sex.fit_transform(df['Sex'])
print("Sex encoded (male→1, female→0)")

# Encode Embarked
LE_embarked = LabelEncoder()
df['Embarked'] = LE_embarked.fit_transform(df['Embarked'])
print("Embarked encoded")

df.head()
```

```python
y = df['Survived']
X = df.drop('Survived', axis='columns')

print(f"X shape: {X.shape}")
print(f"Features: {list(X.columns)}")
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.4, random_state=42
)

print(f"Training: {X_train.shape[0]} samples")
print(f"Testing: {X_test.shape[0]} samples")
```

```python
# Create and train model
model_new = tree.DecisionTreeClassifier(random_state=42)
model_new.fit(X_train, y_train)

# Calculate accuracy
train_score = model_new.score(X_train, y_train)
test_score = model_new.score(X_test, y_test)

print(f"Training Accuracy: {train_score:.3f}")
print(f"Testing Accuracy: {test_score:.3f}")
```

```python
# Train original model (4 features) for comparison
df_original = pd.read_csv("Copy of titanic.csv")
df_original = df_original.drop(columns=['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
df_original['Age'] = df_original.Age.fillna(math.floor(df_original.Age.mean()))
df_original['Sex'] = LabelEncoder().fit_transform(df_original['Sex'])

y_orig = df_original['Survived']
X_orig = df_original.drop('Survived', axis='columns')

X_train_orig, X_test_orig, y_train_orig, y_test_orig = train_test_split(
    X_orig, y_orig, test_size=0.4, random_state=42
)

model_orig = tree.DecisionTreeClassifier(random_state=42)
model_orig.fit(X_train_orig, y_train_orig)
orig_score = model_orig.score(X_test_orig, y_test_orig)

# Compare results
print("COMPARISON RESULTS")
print(f"Original Model (4 features)  : {orig_score:.3f}")
print(f"New Model (7 features)       : {test_score:.3f}")

if test_score > orig_score:
    improvement = (test_score - orig_score) * 100
    print(f"\nIMPROVEMENT: {improvement:.1f}% better!")
else:
    print(f"\nNo improvement. Try different features.")
```

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')  # Warning ignore karein

def predict_survival(pclass, sex, age, sibsp, parch, fare, embarked):
    """
    Predict if a passenger would survive
    """
    # Encode sex
    sex_encoded = 1 if sex.lower() == 'male' else 0
    
    # Encode embarked
    embarked_map = {'S': 0, 'C': 1, 'Q': 2}
    embarked_encoded = embarked_map.get(embarked.upper(), 0)
    
    # Create DataFrame with feature names
    input_data = pd.DataFrame([[pclass, sex_encoded, age, sibsp, parch, fare, embarked_encoded]],
                              columns=['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked'])
    
    # Predict
    prediction = model_new.predict(input_data)
    probability = model_new.predict_proba(input_data)
    
    print(f"Prediction: {'SURVIVED' if prediction[0] == 1 else 'DIED'}")
    print(f"Survival Probability: {probability[0][1]:.2%}")
    
    return prediction[0]

# Example prediction
print("\nExample Prediction:")
predict_survival(3, 'male', 25, 0, 0, 8.05, 'S')
```
